```python
import pandas as pd
import seaborn as sb
```

```python
d=pd.read_csv('titanic.csv')
df=pd.DataFrame(d)
df.describe()
```

```python
'''
Majorly ticket fares were low (<100) 
Most of the children (<10 years) were saved
Aged people (>60) didnt survive much
The rich (>100 fare) also survived in large numbers compared to the poor (<50)
'''
sb.relplot(data=df,x='Age',y='Fare',height=10,hue='Survived')
```

```python
'''
Majorly ticket fares were low (<100)
2 males who bought the most expensive tickets (>300) survived
Most of the aged people (>60) were male
Many males paid low fares
Middle aged females (20 to 50) bought tickets of different fares (20 to 250)
'''
sb.relplot(data=df,x='Age',y='Fare',height=10,hue='Sex')
```

```python
'''
Around 700 tickets within fare of 50 of which 200 females and 500 males
Around 100 tickets in the range of 50 to 100 of which count of males ~ females ~50
Ticket fares above 150 were bought in negligible quantity
There were more males than females  on the ship
'''
sb.histplot(data=df,x=df.Fare,bins=10,hue='Sex')
```

```python
'''
Around 700 tickets within fare of 50 of which ~200 survived and ~500 died
Around 100 tickets in the range of 50 to 100 of which count of many survived
Ticket fares above 150 were bought in negligible quantity
There were less survivors
'''
sb.histplot(data=df,x=df.Fare,bins=10,hue='Survived',kde=True)
```

```python
'''
There were less survivors in all age groups except <10
'''
sb.histplot(data=df,x=df.Age,bins=10,hue='Survived',kde=True)
```

```python
'''
median age of female survivors > female dead
more males of lower age survived than dead (based on 1st quantile)
'''
sb.boxplot(data=df,x='Sex',y='Age',hue='Survived')
```

```python
'''
Median age of survivors ~ median age of dead ~ 29
'''
sb.boxplot(data=df,x='Survived',y='Age')
```

```python
from nltk.tokenize import *
from nltk.corpus import stopwords
import string
from nltk import *
```

```python
file=open("cv011_13044.txt")
text=file.read()
text
```

```python
# tokenizing into sentences
tok_sent=sent_tokenize(text)
tok_sent
```

```python
# tokenizing sentences into words
tok_sent_words=[]
for sentence in tok_sent:
    tok_sent_words.append(word_tokenize(sentence))
for ws in tok_sent_words:
    print(ws)
```

```python
# pos tagging
for ws in tok_sent_words:
    print(pos_tag(ws))
    print()
```

```python
stop_words=stopwords.words('english')
punctuations=list(string.punctuation)+['``','--']
```

```python
# stopwords and punctuations removal
for ws in tok_sent_words:
    for i in stop_words:
        while i in ws:
            ws.remove(i)
    for i in punctuations:
        while i in ws:
            ws.remove(i)
    print(ws)
```

```python
# lemmatize and stemming
stemmer=stem.PorterStemmer()
lemmatizer=stem.WordNetLemmatizer()
for ws in tok_sent_words:
    for w in ws:
        print(w,"=",lemmatizer.lemmatize(w)," ; ",stemmer.stem(w))
    print()
```

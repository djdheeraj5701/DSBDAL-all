```python
from nltk.tokenize import *
from nltk.corpus import stopwords
import string
from nltk import *
from collections import *
from math import *
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
class TFxIDF:
    def __init__(self):
        self.tf=defaultdict(dict)
        self.idf=defaultdict(float)
    def appendTF(self,word,count,document):
        self.tf[word][document]=count
    def appendIDF(self,word,count):
        self.idf[word]=count
    
```

```python
tfxidf=TFxIDF()
```

```python
stop_words=stopwords.words('english')
punctuations=list(string.punctuation)+['``','--',"''"]
for document in range(1,26):
    file=open("Ass7_Dataset/"+str(document))
    text=file.read()
    tok_sent=sent_tokenize(text)
    tok_sent_words=[]
    for sentence in tok_sent:
        sentence=sentence.lower()
        tok_sent_words.append(word_tokenize(sentence))
    final_words=[]
    for ws in tok_sent_words:
        for i in stop_words:
            while i in ws:
                ws.remove(i)
        ws2=[]
        for i in ws:
            for j in list('0123456789')+punctuations:
                if j in i:
                    break
            else:
                ws2.append(i)
        final_words+=ws2
    counts=Counter(final_words)
    for i in counts:
        tfxidf.appendTF(i,counts[i]/len(final_words),document)
#         print(i,counts[i]/len(final_words),document)
for word in tfxidf.tf:
    dfcount=0
    for document in range(1,26):
        if document in tfxidf.tf[word]:
            dfcount+=1
    idfcount=25/(dfcount)
    tfxidf.appendIDF(word,log(idfcount))
#     print(word,idfcount)
    
```

```python
importanceY=defaultdict(list)
for word in tfxidf.tf:
    for document in range(1,26):
        if document in tfxidf.tf[word]:
            TF,IDF=tfxidf.tf[word][document],tfxidf.idf[word]
        else:
            TF,IDF=1,0
        importanceY[word].append(TF*IDF)
for word in tfxidf.tf:
    setY=set(importanceY[word])
    setY.remove(0) if 0 in setY else None
    if  min(setY)<0.05 or max(setY)<0.1:
        continue
    print(word,importanceY[word])
    plt.plot(
        np.array(list(range(1,26))),
        np.array(importanceY[word])
    )
    try:
        plt.title(word)
        plt.xticks(list(range(1,26)))
        plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
        plt.savefig('imgs/'+word+'.png')
    except:
        pass
    plt.clf()
```
